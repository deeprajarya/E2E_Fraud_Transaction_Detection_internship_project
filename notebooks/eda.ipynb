Lets determine number of Fraud cases in the dataset

 0 : valid

 1 : fraud

```python
fraud = df[df['Class']==1]
fraud.head()
```

```python
valid = df[df['Class']==0]
valid.head()
```

```python
print(f"Total Fraud Transaction :",len(fraud))
print(f"Total Valid Transaction :",len(valid))
```

```python
# fraction of fraud and valid cases

fraction = len(fraud)/len(valid)
```

```python
total_transaction = len(fraud)+len(valid)
print(f"Total Transaction :",total_transaction)

frac_percnt = len(fraud)/total_transaction

print(round(frac_percnt*100,2),"% transaction are fraud in the dataset")
```

```python
# Lets check fraudent transaction

print(fraud.Amount.describe())
print("\nTotal amount of fraud transaction :",fraud.Amount.sum())
```

```python
# Lets check valid transaction

print(valid.Amount.describe())
print("\nTotal amount of valid transaction :",valid.Amount.sum())
```
